```python
import matplotlib.pyplot as plt
```

# visualização de dados em Python


```python
x = [1,3,5,7,9]
y = [214, 222, 215, 218, 206]

titulo = "Gráfico de dispersão"
eixox = "Eixo X"
eixoy = "Eixo Y"

#legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.plot(x, y, color="#00008B", linestyle="-")
plt.scatter(x, y, label="pontos", color="r", marker=".", s=150)
plt.legend()
plt.show()
```
