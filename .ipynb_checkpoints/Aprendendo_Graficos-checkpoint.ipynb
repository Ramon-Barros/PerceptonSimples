```python
import matplotlib.pyplot as plt
```

# visualização de dados em Python


```python
x = [1,3,5,7,9]
y = [2,3,7,1,0]

titulo = "Gráfico de dispersão"
eixox = "Eixo X"
eixoy = "Eixo Y"

#legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.plot(x, y, color="r", linestyle="-")
plt.scatter(x, y, label="pontos", color="k", marker=".", s=200)
plt.legend()
plt.show()
```
